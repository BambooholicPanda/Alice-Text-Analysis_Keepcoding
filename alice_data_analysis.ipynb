```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

I have exported the dictionary of words and word count of the Alice and Wonderland text through my code on VSC and I have uploaded it here on GitHub. The table contains 2084 rows (with the 2084 words that appear on the text, excluding stopwords) and 3 columns that display the word, the word count and the word's length, respectively. You can see a preview of the table right below this comment.

```python
alice_table = pd.read_csv("https://raw.githubusercontent.com/BambooholicPanda/Alice-Text-Analysis_Keepcoding/main/alicia_tabla.csv")

display(alice_table)
```

This table right below shows the words that appear the least on the text.

```python
alice_table.sort_values(by=['Count']).head(20)
```

This table right below shows the words that appear the most on the text. The most used word is "alice". It's funny that the second most used word is "queen", but there's a gap of 322 uses among them.

```python
alice_table.sort_values(by=['Count']).tail(20).iloc[::-1]
```

This bar graph right below shows the word repetition frequency. It is clear that the greatest tendency is that a word is used only once. We can also see that words don't tend to repeat themselves more than 10 times.

```python
alice_table['Count'].value_counts().plot(kind='bar')
```

The table below shows the shortest words in length.

```python
alice_table.sort_values(by=['Length']).head(20)
```

The table below shows the longest words in length. We can appreciate that most of these words are compound words. 

```python
alice_table.sort_values(by=['Length']).tail(20).iloc[::-1]
```

This bar graph below shows the word length frequency. We can see that the most common words are those which have a length of 6 characters.

```python
alice_table['Length'].value_counts().plot(kind='bar')
```

Thanks to data analysis, we can compare, for example, the frequency of appearance of words like "queen" and "king". The queen wins!

```python
search = alice_table.loc[alice_table['Word'].isin(["queen", "king"])]
barplot = plt.bar(x=search["Word"], height=search["Count"])
plt.bar_label(barplot)
plt.show()
```

I also wanted to check the frequency of appearance of the words represent Alice in Wonderland to me. I picked "alice", "rabbit", "drink", "nonsense", "queen", "cheshire", "hatter" and "mad" and here you can see the results. They do play an important role in the story, and the bar graph proves it.

```python
search = alice_table.loc[alice_table['Word'].isin(["alice", "hatter", "nonsense", "drink", "rabbit", "mad", "queen", "chess", "cheshire", "hole"])]
barplot = plt.bar(x=search["Word"], height=search["Count"], fc="pink", ec="black")
plt.bar_label(barplot)
plt.show()
```
