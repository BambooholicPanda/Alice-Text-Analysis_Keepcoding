```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Leo el archivo y muestro la tabla completa

```python
alice_table = pd.read_csv("https://raw.githubusercontent.com/BambooholicPanda/Alice-Text-Analysis_Keepcoding/main/alicia_tabla.csv")

display(alice_table)
```

Valores mas pequeños de count

```python
alice_table.sort_values(by=['Count']).head(20)
```

Valores mas grandes de count

```python
alice_table.sort_values(by=['Count']).tail(20).iloc[::-1]
```

Frecuencia de los valores de count

```python
alice_table['Count'].value_counts().plot(kind='bar')
```

Palabras mas cortas

```python
alice_table.sort_values(by=['Length']).head(20)
```

Palabras mas largas

```python
alice_table.sort_values(by=['Length']).tail(20).iloc[::-1]
```

Frecuencia de longitud de las palabras

```python
alice_table['Length'].value_counts().plot(kind='bar')
```

Comparacion de palabras especificas

```python
alice_table.loc[alice_table['Word'].isin(["king", "queen"])].value_counts().plot(kind='bar')
```
